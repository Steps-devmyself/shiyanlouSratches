```python
#2.在序列循环中，同时获得索引位置和对应的值
```

```python
words=['ticc','tac','toe','what']
```

```python
for i,v in enumerate(words):
    print(i,v)
```

```python
#对于字典也可以用，返回的是key的索引位置
```

```python
for i,v in enumerate(tel):
    print(i,v)
```

```python
# 同时循环两个或更多个序列，使用zip()整体打包
```

```python
questions = ['name', 'quest', 'favorite color']
```

```python
answers = ['lancelot', 'the holy grail', 'blue']
```

```python
for q,a in zip(questions,answers):
    print(q,a)
```

```python
# 多个循环序列长度不一致时,以短的为准
```

```python
for q,a,t in zip(questions,answers,words):
    print(q,a,t)
```

```python
# 逆向循环(reversed),排序循环(sorted),这两个函数都不是inplace，而是原来的列表不变，返回一个新列表
```

```python
for i in reversed(words):
    print(i)
```

```python
for i in sorted(questions):
    print(i)
```

```python
# 在循环内部修改正在遍历的序列，先制作序列的副本,序列在循环的时候不会隐式地创建副本，也就是不会陷入死循环
```

```python
# 也就是说words[:]这也可以创建副本
```

```python
#像这样就会产生死循环
for w in words:
    print(w)
    if len(w)>3:
        words.insert(0,w)
```

```python
for w in words[:]:
    if len(w)>3:
        words.insert(0,w)
```
